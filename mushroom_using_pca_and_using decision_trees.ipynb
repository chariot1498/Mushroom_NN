```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#This command gives us the mushrooms dataset as a pandas dataframe
ds = pd.read_csv('mushrooms.csv')
ds.info()
ds
```

```python
#Using labelencoder to convert characters in columns to numbers.
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
ds = ds.apply(le.fit_transform)
```

```python
#Deciding the split for training and testing data
split = int(0.8 * ds.shape[0])
```

```python

ytrain = ds[['class']][:split]
ytest = ds[['class']][split:]
```

```python
ytrain
```

```python
ds = ds.drop('class',1)
```

```python
xtrain = ds[:split]
xtest = ds[split:]
```

```python
xtrain
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtrain.dropna
```

```python
xtrain.shape
```

```python
#Using decision trees and random forest to have a classification on basis of all features.
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
```

```python
dt.score(xtest,ytest)
```

```python
rf = RandomForestClassifier(n_estimators=10)
rf.fit(xtrain, ytrain)

rf.score(xtest, ytest)
```

```python
#Now we are going to apply pca and see how much accuracy can we get from less features. You can change the number of features you want to keep in the n_components in PCA Classifier     
df = pd.read_csv('mushrooms.csv')
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=10)
```

```python
#Done for trsnsforming characters in features to numbers
df = df.apply(le.fit_transform)
```

```python
dd = df[['class']]
```

```python
#'class' is a classification feature we don't want to apply pca with it we want to remove it . We only want features that are pred func is going to take   
df = df.drop('class',1)
```

```python
#This command converts our 22 features to 10 features
df = pca.fit_transform(df)
```

```python
df
```

```python
#Splitting for testing and training data
split = int(0.8 * df.shape[0])
```

```python
ytrain = dd[['class']][:split]
ytest = dd[['class']][split:]
```

```python
xtrain = df[:split]
xtest = df[split:]
```

```python
xtrain
```

```python
#Using decision trees and random classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
```

```python
dt.score(xtest,ytest)
```

```python
rf = RandomForestClassifier(n_estimators=100)
rf.fit(xtrain, ytrain)

rf.score(xtest, ytest)
```
