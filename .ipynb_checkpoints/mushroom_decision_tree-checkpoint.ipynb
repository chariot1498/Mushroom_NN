```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#This command reads a csv file to a pandas dataframe
ds = pd.read_csv('mushrooms.csv')
ds.info()
```

```python
ds.head()
```

```python
#Did this to drop rows having Nan value in any column
ds.dropna
ds.shape
ds = ds.reset_index(drop=True)
```

```python
#Selected important features for classification of this dataset . 
df = ds[['odor','spore_print_color','habitat','population','cap_color','stalk_surface_above_ring']]
dd = ds[['type']]
df
```

```python
#Didn't wanted to use labelencoder so manually coded each feature from a character representation to a numeric representation
def odor_int(s):
    if s == 'a':
        return 0
    elif s == 'l':
        return 1
    elif s == 'c':
        return 2
    elif s == 'y':
        return 3
    elif s == 'f':
        return 4
    elif s == 'm':
        return 5
    elif s =='n':
        return 6
    elif s == 'p':
        return 7
    elif s == 's':
        return 8
    else :
        return s
df.odor = df.odor.map(odor_int)
```

```python
def spore_int(s):
    if s == 'k':
        return 0
    elif s == 'n':
        return 1
    elif s == 'b':
        return 2
    elif s == 'h':
        return 3
    elif s == 'r':
        return 4
    elif s == 'o':
        return 5
    elif s == 'u':
        return 6
    elif s == 'w':
        return 7
    elif s == 'y':
        return 8
    else :
        return s
df.spore_print_color = df.spore_print_color.map(spore_int)
```

```python
def habitat_int(s):
    if s == 'g':
        return 0
    if s == 'l':
        return 1
    if s == 'm':
        return 2
    if s == 'p':
        return 3
    if s == 'u':
        return 4
    if s == 'w':
        return 5
    if s == 'd':
        return 6
    return s
df.habitat = df.habitat.map(habitat_int)
```

```python
def pop_int(s):
    if s == 'a':
        return 0
    if s == 'c':
        return 1
    if s == 'n':
        return 2
    if s == 's':
        return 3
    if s == 'v':
        return 4
    if s == 'y':
        return 5
    return s
df.population = df.population.map(pop_int)
```

```python
def cap_color_int(s):
    if s == 'n':
        return 0
    if s == 'b':
        return 1
    if s == 'c':
        return 2
    if s == 'g':
        return 3
    if s == 'r':
        return 4
    if s == 'p':
        return 5
    if s == 'u':
        return 6
    if s == 'e':
        return 7
    if s == 'w':
        return 8
    if s == 'y':
        return 9
    return s
df.cap_color = df.cap_color.map(cap_color_int)
```

```python
def stalk_surface_above_ring_int(s):
    if s == 'f':
        return 0
    if s == 'y':
        return 1
    if s == 'k':
        return 2
    if s == 's':
        return 3
    return s
df.stalk_surface_above_ring = df.stalk_surface_above_ring.map(stalk_surface_above_ring_int)
```

```python
def type_int(s):
    if s == 'p':
        return 0
    if s == 'e':
        return 1
    return s
dd.type = dd.type.map(type_int)
```

```python
#deciding slpit for dividing data into testing and trainning
split = int(0.75 * df.shape[0])
```

```python
df.shape,dd.shape
```

```python
xtrain = df[:split]
xtest = df[split:]
ytrain = dd[:split]
ytest = dd[split:]
```

```python
#Calling sklearn library for decision trees and random forest ensemble though if you want you can code it from scratch
import sklearn
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
```

```python
dt.score(xtest,ytest)
```

```python
#Random forest is a ensemble model it is using 100 decision trees and takes the best score it gets on the dataset.
rf = RandomForestClassifier(n_estimators=100)
rf.fit(xtrain, ytrain)

rf.score(xtest, ytest)
```
